```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
```

```python
df = pd.read_csv("FuelConsumptionEdit2.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
```

```python
cdf.head()
```

```python
cdf.hist()
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS, color='blue')
plt.xlabel('ENGINESIZE')
plt.ylabel('CO2EMISSIONS')
plt.show
```

```python
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, color='green')
plt.xlabel('CYLINDERS')
plt.ylabel('CO2EMISSIONS')
plt.show
```

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS, color='red')
plt.xlabel('FUELCONSUMPTION_COMB')
plt.ylabel('CO2EMISSIONS')
plt.show
```

```python
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = df[~msk]
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='yellow')
plt.xlabel('TRAIN ENGINESIZE')
plt.ylabel('TRAIN CO2EMISSIONS')
plt.show
```

```python
from sklearn import linear_model
regresi = linear_model.LinearRegression()
train_x = np.asanyarray(train[['ENGINESIZE']])
train_y = np.asanyarray(train[['CO2EMISSIONS']])
regresi.fit(train_x, train_y)

print('Koefisien : ', regresi.coef_)
print('Intercept : ', regresi.intercept_)
```

```python
plt.scatter(train.ENGINESIZE, train.CO2EMISSIONS, color='green')
plt.plot(train_x,regresi.coef_[0][0]*train_x + regresi.intercept_[0], '-r')
plt.xlabel('TRAIN ENGINESIZE')
plt.ylabel('TRAIN CO2EMISSIONS')
plt.show
```

```python
test_x = np.asanyarray(test[['ENGINESIZE']])
test_y = np.asanyarray(test[['CO2EMISSIONS']])
```

```python
test_y = regresi.predict(test_x)
```

```python
test_y
```

```python
prediksi_y = regresi.predict([[3]])
prediksi_y
```

```python
print("Nilai Mean Absolute Error : %.2f" % np.mean(np.absolute(prediksi_y - test_y)))
```
